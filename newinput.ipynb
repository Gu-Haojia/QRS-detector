```python
annotation = wfdb.rdann('data_records/generated_data/x001', 'aux')
#annotation.__dict__
```

```python
devices=['SOT','FAROS','NEXUS','HEXOSKIN']
participants=['x001','x002','x003','x004','x005','x006','x007','x008','x009','x010','x011','x012','x013']
qrs_widths=20
plot=False
scale=300
```

```python
#导入数据
for device in devices:
    #新建一个空列表，用于存放数据
    data = []

    for participant in participants:
        print('proccessing '+device+' '+participant)

        topdata = {}
        records = wfdb.rdrecord('data_records/generated_data/'+participant)
        annotation = wfdb.rdann('data_records/generated_data/'+participant, 'atr')
        activity = wfdb.rdann('data_records/generated_data/'+participant, 'aux')
        atr=annotation.__dict__['sample']
        signal = records.p_signal[:,devices.index(device)]
        
        """
        #signal以0为中心，缩放到-5到5之间
        #找到绝对值最大的0.4%的所有数据
        max_value = np.percentile(np.abs(signal),99.6)
        #计算最大的0.4%数据的平均值
        mean_value = np.mean(np.abs(signal[signal>max_value]))
        #将数据缩放到-5到5之间
        signal = signal/mean_value*5
        """

        #将atr的标记为中心，前后qrs_widths/2的数据作为qrs_lable
        qrs_lable =  np.zeros(len(signal),dtype=int)
        for i in atr:
            if i-qrs_widths/2>0:
                start=i-qrs_widths/2
            else:
                start=0
            if i+qrs_widths/2<len(signal):
                end=i+qrs_widths/2
            else:
                end=len(signal)

            start=int(start)
            end=int(end)
            #在qrs_lable中添加start到end的1序列
            qrs_lable[start:end]=1

        #生成r的one-hot布尔序列
        rr = np.zeros(len(signal),dtype=bool)
        rr[atr]=1


        raw = {}
        raw['ecg'] = signal
        raw['r'] = atr
        raw['qrs'] = qrs_lable
        raw['participant'] = participant
        raw['rr'] = rr
        #全部数据
        topdata['raw'] = raw

        #按活动划分数据
        #获取活动的开始和结束时间
        FAROS_Marker_Rest=activity.__dict__['sample'][0]
        FAROS_Marker_Walking=activity.__dict__['sample'][1]
        Manual_Walking=activity.__dict__['sample'][2]
        FAROS_Marker_2_Back=activity.__dict__['sample'][3]
        Manual_Running=activity.__dict__['sample'][4]

        rest_start = FAROS_Marker_Rest
        rest_end = min(FAROS_Marker_Walking,Manual_Walking)-1
        walking_start = max(Manual_Walking,FAROS_Marker_Walking)
        walking_end = FAROS_Marker_2_Back-1
        back_start = FAROS_Marker_2_Back
        back_end = Manual_Running-1
        running_start = Manual_Running
        running_end = len(signal)

        #将数据按活动划分
        rest = {}
        rest['ecg'] = signal[rest_start:rest_end]
        rest['r'] = atr[(atr>rest_start) & (atr<rest_end)]-rest_start
        rest['qrs'] = qrs_lable[rest_start:rest_end]
        rest['rr']=rr[rest_start:rest_end]
        rest['participant'] = participant
        topdata['rest'] = rest

        #打印rest规格
        print('rest of '+device+' '+participant+' is '+str(len(rest['ecg'])))

        #随机展示scale长度的rest数据
        if plot:
            plt.figure(figsize=(10,6))
            plt.plot(rest['ecg'][:scale])
            plt.plot(rest['qrs'][:scale])
            #在rest数据中标记r波
            for i in rest['r']:
                if i<scale:
                    plt.axvline(x=i,color='r')
            plt.title(device+participant+'_rest')
            plt.show()

        walking = {}
        walking['ecg'] = signal[walking_start:walking_end]
        walking['r'] = atr[(atr>walking_start) & (atr<walking_end)]-walking_start
        walking['qrs'] = qrs_lable[walking_start:walking_end]
        walking['rr']=rr[walking_start:walking_end]
        walking['participant'] = participant
        topdata['walking'] = walking

        #打印walking规格
        print('walking of '+device+' '+participant+' is '+str(len(walking['ecg'])))

        #随机展示scale长度的walking数据
        if plot:
            print('walking')
            plt.figure(figsize=(10,6))
            plt.plot(walking['ecg'][:scale])
            plt.plot(walking['qrs'][:scale])
            #在walking数据中标记r波
            for i in walking['r']:
                if i<scale:
                    plt.axvline(x=i,color='r')
            plt.title(device+participant+'_walking')
            plt.show()

        back = {}
        back['ecg'] = signal[back_start:back_end]
        back['r'] = atr[(atr>back_start) & (atr<back_end)]-back_start
        back['qrs'] = qrs_lable[back_start:back_end]
        back['rr']=rr[back_start:back_end]
        back['participant'] = participant
        topdata['back'] = back

        #打印back规格
        print('back of '+device+' '+participant+' is '+str(len(back['ecg'])))

        #随机展示scale长度的back数据
        if plot:
            plt.figure(figsize=(10,6))
            plt.plot(back['ecg'][:scale])
            plt.plot(back['qrs'][:scale])
            #在back数据中标记r波
            for i in back['r']:
                if i<scale:
                    plt.axvline(x=i,color='r')
            plt.title(device+participant+'_back')
            plt.show()

        running = {}
        running['ecg'] = signal[running_start:running_end]
        running['r'] = atr[(atr>running_start) & (atr<running_end)]-running_start
        running['qrs'] = qrs_lable[running_start:running_end]
        running['rr']=rr[running_start:running_end]
        running['participant'] = participant
        topdata['running'] = running

        #打印running规格
        print('running of '+device+' '+participant+' is '+str(len(running['ecg'])))

        #随机展示scale长度的running数据
        if plot:
            plt.figure(figsize=(10,6))
            plt.plot(running['ecg'][:scale])
            plt.plot(running['qrs'][:scale])
            #在running数据中标记r波
            for i in running['r']:
                if i<scale:
                    plt.axvline(x=i,color='r')
            plt.title(device+participant+'_running')
            plt.show()

        data.append(topdata)

    #将数据存入dataframe
    data = pd.DataFrame(data)
    #pickle存储
    data.to_pickle('./pickle/data_'+device+'.pkl')

        
```
